```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

housing = pd.read_table('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data',header=None, sep='\s+',
                       names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE","DIS", 
                                             "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"])
```

```python
from scipy.stats import skew

stats = housing.describe().T
stats['skew'] = skew(housing)
stats
```

```python
import random

np.random.seed(12)

sample1pct = housing.sample(5)
sample1pctstats = pd.DataFrame({
                    'Sample Mean': sample1pct.mean(),
                    'Diff from Mean': sample1pct.mean() - stats['mean'],
                    'Norm diff fr M': (sample1pct.mean() - stats['mean'])/stats['std']
})
sample1pctstats
```

```python
stats = sample1pct.describe().T
stats['skew'] = skew(sample1pct)
stats
```

```python
sample1pct
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
sample_sc = scaler.fit_transform(sample1pct)
sample_sc_df = pd.DataFrame(sample_sc, columns=housing.columns)
```

```python
sample_sc_df
```

```python
mask = np.zeros_like(sample_sc_df)
mask[np.triu_indices_from(mask, 0)] = False
with sns.axes_style("white"):
    plt.figure(figsize=(12, 4))
    ax = sns.heatmap(sample_sc_df, mask=mask, square=True, annot=True,
                     cmap='RdBu', fmt='+.3f')
    plt.xticks(rotation=45, ha='center')
```

```python
fig = plt.figure(figsize=(20,12))
for i, col in enumerate(sample_sc_df.columns):
    fig.add_subplot(4,4,i+1)
    sns.distplot(sample_sc_df[col], label=col)
    plt.axvline(sample_sc_df[col].mean(), c='red')
    plt.axvline(sample_sc_df[col].median(), c='black')
    plt.legend()
```
